```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x=np.arange(0,10)
y=[3,4,5,6,7,8,9,10,11,23]
datax=pd.DataFrame(x)
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.show()
```

```python
x=6*np.random.rand(250,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(250,1)
plt.scatter(x,y,color='g')
plt.xlabel('x points')
plt.ylabel('y points')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg_Model1=LinearRegression()
```

```python
reg_Model1.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
score1=r2_score(y_test,reg_Model1.predict(x_test))
print(score1)
```

```python
plt.plot(x_train,reg_Model1.predict(x_train),color='y')
plt.scatter(x_train,y_train)
plt.xlabel("x values")
plt.xlabel("y values")
```

polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
p=PolynomialFeatures(degree=2,include_bias=True)
x_train_p=p.fit_transform(x_train)
x_test_p=p.transform(x_test)
```

```python
x_train_p
```

```python
x_test_p
```

```python
reg_Model2=LinearRegression()
reg_Model2.fit(x_train_p,y_train)
y_predict2=reg_Model2.predict(x_test_p)
score2=r2_score(y_test,y_predict2)
```

```python
print(score2)
```

```python
print(reg_Model2.coef_)
```

```python
reg_Model3=LinearRegression()
reg_Model3.fit(x_train_p,y_train)
y_predict3=reg_Model3.predict(x_test_p)
score3=r2_score(y_test,y_predict3)
```

```python
print(score3)
```
