```python
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = bs.data
y = bs.target
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.02 ,random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(xtrain,ytrain)
ypred = knn.predict(xtest)

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
dec = DecisionTreeClassifier()
rand = RandomForestClassifier()
dec.fit(xtrain,ytrain)
rand.fit(xtrain,ytrain)
ypred2 = dec.predict(xtest)
ypred3 = rand.predict(xtest)
```

```python
accuracy = accuracy_score(ytest,ypred)
print("Accuracy of KNN before PCA:",accuracy)

print("Accuracy of Decsion Tree Classifier before PCA:", accuracy_score(ytest,ypred2))
print("ACcuracy of Random Forest Classifier before PCA:",accuracy_score(ytest,ypred3))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,fmt='d',cmap="Blues",xticklabels=["Predicted 0","Predicted 1"], yticklabels=["Actual 0","Actual 1"])
plt.title("Confusion Matrix for KNN")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(ytest,ypred2),annot=True ,cmap="Greens",xticklabels=["Predicted 0","Predicted 1"],yticklabels=["Actual 0","Actual 1"])
plt.title("Confusion Matrix for Decision Tree Classifier")
plt.show()
```

```python
plt.figure(figsize =(8,6))
sns.heatmap(confusion_matrix(ytest,ypred3),annot=True,fmt='d',cmap="magma",xticklabels=["Predicted 0","Predicted 1"],yticklabels=["Actual 0", "Actual 1"])
plt.title("Confusion Matrix for Random Forest Classifier")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(x)
pca_bs = PCA(n_components=2)
X_pca = pca_bs.fit_transform(X)
bs_pca =pd.DataFrame(data =X_pca)
bs_pca["target"] =y
```

```python
X =X_pca
Y = bs_pca["target"]

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X,Y)
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)
ypred5 = knn.predict(xtest)

dec = DecisionTreeClassifier()
rand= RandomForestClassifier()
dec.fit(xtrain,ytrain)
rand.fit(xtrain,ytrain)
ypred6 = dec.predict(xtest)
ypred7= rand.predict(xtest)
```
